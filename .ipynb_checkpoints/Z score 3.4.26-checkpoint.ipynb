```python

import matplotlib as mpl
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd
import plotly.express as px
import pandas as pd

mpl.rcParams['figure.figsize'] = (10, 8)
mpl.rcParams['axes.grid'] = False

dataset_name = "3.4.26"
```

```python
df = pd.read_csv(dataset_name +".csv")
df.head()
```

```python
fig = px.line(df.reset_index(), x='datetime', y='source_addr', title='Whole Dataset')

fig.update_xaxes(
    rangeslider_visible=True,
)
fig.show()
```

```python
df = df[(df['datetime'] >= '2019-07-19') & (df['datetime'] <= '2021-10-19')]
df.shape
```

## Z score

```python

plt.hist(df["source_addr"], bins = 20, rwidth = 0.8)
plt.show()
```

```python
df["zscore"] = (df.source_addr - df.source_addr.mean()) / df.source_addr.std()
df
```

```python
df[(df["zscore"] > 3) | (df["zscore"] < -3)]
```

## Detect Anomaly

```python
df['anomaly'] = df.zscore.apply(lambda x: 'Yes' if((x > 3) | (x < -3)) else 'No')
```

```python
df.loc[df["anomaly"]== "Yes"]
```

```python
df.loc[df["anomaly"]== "No"]
```

```python
fig = px.scatter(df.reset_index(), x='datetime', y='source_addr', color='anomaly', title='Anomaly Detection')

fig.update_xaxes(
    rangeslider_visible=True
    
)
fig.show()
```
